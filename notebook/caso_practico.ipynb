```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/home/usuario/caso-practico/datasets/Leads.csv')
print(len(df))
df.head().T
```

```python
df.dtypes
```

```python
replacer = lambda str: str.lower().str.replace(' ', '_')
df.columns = replacer(df.columns.str)

for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = replacer(df[col].str)

df.head().T
```

```python
df.do_not_email = (df.do_not_email == 'yes').astype(int)
df.do_not_call = (df.do_not_call == 'yes').astype(int)
df.search = (df.search	 == 'yes').astype(int)
df.magazine = (df.magazine	 == 'yes').astype(int)
df.newspaper_article = (df.newspaper_article == 'yes').astype(int)
df.x_education_forums = (df.x_education_forums == 'yes').astype(int)
df.newspaper = (df.newspaper == 'yes').astype(int)
df.digital_advertisement = (df.digital_advertisement == 'yes').astype(int)
df.through_recommendations = (df.through_recommendations == 'yes').astype(int)
df.receive_more_updates_about_our_courses = (df.receive_more_updates_about_our_courses == 'yes').astype(int)
df.update_me_on_supply_chain_content = (df.update_me_on_supply_chain_content == 'yes').astype(int)
df.get_updates_on_dm_content = (df.get_updates_on_dm_content == 'yes').astype(int)
df.i_agree_to_pay_the_amount_through_cheque = (df.i_agree_to_pay_the_amount_through_cheque == 'yes').astype(int)
df.a_free_copy_of_mastering_the_interview = (df.a_free_copy_of_mastering_the_interview == 'yes').astype(int)
	
df.head().T
```

```python
replacerSelect = lambda str: str.replace('select', np.isnan)
df.columns = replacerSelect(df.columns.str)

for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = replacerSelect(df[col].str)

df.head().T
```

```python
df = df.fillna('desconocido')
df.T
```

```python
from sklearn.compose import make_column_selector as selector

categorical_columns_selector = selector(dtype_include=object)
categorical_columns = categorical_columns_selector(df)


numerical_columns_selector = selector(dtype_exclude=object)
numerical_columns = numerical_columns_selector(df)

categorical_columns
numerical_columns
```

```python
from sklearn.model_selection import train_test_split

df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)

df_train, df_validation = train_test_split(df_train_full, test_size=0.30, random_state=1)

y_train = df_train.converted.values
y_validation = df_validation.converted.values

del df_train['converted']
del df_validation['converted']

df_train_full.head().T
```

```python
global_mean = df_train_full.converted.mean()
round(global_mean, 3)
```

```python
from sklearn.metrics import mutual_info_score

calculate_mi = lambda col:mutual_info_score(col, df_train_full.converted)

df_mi = df_train_full[categorical_columns].apply[calculate_mi]
df_mi = df_mi.sort_values(ascending = False).to_frame(name='MI')
df_mi
```

```python
numerical = ['lead_number',
 'do_not_email',
 'do_not_call',
 'totalvisits',
 'total_time_spent_on_website',
 'page_views_per_visit',
 'search',
 'magazine',
 'newspaper_article',
 'x_education_forums',
 'newspaper',
 'digital_advertisement',
 'through_recommendations',
 'receive_more_updates_about_our_courses',
 'update_me_on_supply_chain_content',
 'get_updates_on_dm_content',
 'asymmetrique_activity_score',
 'asymmetrique_profile_score',
 'i_agree_to_pay_the_amount_through_cheque',
 'a_free_copy_of_mastering_the_interview']
```

```python
train_dict = df_train[categorical_columns + numerical].to_dict(orient='records')
dict(sorted(train_dict[0].items()))
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
x_train = dv.transform(train_dict)
x_train[0]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear')
model.fit(x_train, y_train)
```

```python
val_dict = df_validation[categorical_columns + numerical].to_dict(orient = 'records')
x_validation = dv.transform(val_dict)
y_pred = model.predict_proba(x_validation)[:, 1]
y_pred
```

```python
converted = y_pred >= 0.5
converted
```

```python
round((y_validation == converted).mean(), 3)
```
